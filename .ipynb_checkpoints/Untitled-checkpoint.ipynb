```python
import dash
import dash_core_components as dcc
import dash_html_components as html
import dash_bootstrap_components as dbc
from dash.dependencies import Output, Input, State
import iexfinance
from dateutil.relativedelta import relativedelta
import plotly.graph_objs as go
import datetime
import pandas as pd
import requests
from iexfinance.stocks import get_historical_data

start = datetime.datetime.today() - relativedelta(years=5)
end = datetime.datetime.today()


colors = {
    'background': 'black ',
    'background-body':'black ',
    'text': '#7FDBFF'
}
df = pd.read_csv('patients-maroc.csv')
columns = df.columns


app = dash.Dash(__name__, external_stylesheets=[dbc.themes.BOOTSTRAP], meta_tags=[{"name": "viewport", "content": "width=device-width, initial-scale=1"}])
app.title ='Patients Morocco Covid 19'

app.layout = html.Div(style={'backgroundColor': colors['background-body'],'max-width': '100%', 'overflow-x': 'hidden'}, children=[
    #header
  
    dbc.Row([
          html.Div(html.Div(
                    html.Div(
                        html.Img(src='./assets/UM6P.png'),
                        )
                    ), className="col-md-4 text-center"),
                html.Div(html.Div(
                    html.H2("Patients Morocco Covid 19", style={'color': "white"})
                ), className="col-md-4 text-center"),
                html.Div(html.Div(children=[
                    html.Img(src="/assets/Coronavirus-Masque-covid-19.jpg"),
                ]), className="col-md-4 text-center"),  
            ], justify="center", align="center", className="h-50", style={'backgroundColor': "black"},
        ),

     #body
     
         dbc.Row([
             dbc.Col(
                    html.Div(children=[
                        html.Hr(),
                        html.Label('choose chart',style={'color': "red"}),
                        html.Div([
                           dbc.RadioItems(id="plot_name",
                                options=[
                                    {'label':'Scatter-plot','value':'Scatter-plot'},
                                    {'label':'Line-chart','value':'Line-chart'},
                                    {'label':'Bar-chart','value':'Bar-chart'}],
                                value = 'Bar-chart',
                                labelClassName="date-group-labels",
                                labelCheckedClassName="date-group-labels-checked",
                                className="date-group-items",
                                inline=True,
                                )]),    
                    ],className="col-md-12 text-center",style={'margin-top': '20px','margin-bottom': '20px'})
                ), ]),
        dbc.Row([
              dbc.Col(
                    html.Div(children=[
                        html.Div([  
                            dcc.Dropdown(id='xaxis',
                                         options=[{'label':i,'value':i} for i in columns],
                                         value= columns[0])],style={'width':'30%','display':'inline-block'}),
                        html.Div([
                            dcc.Dropdown(id='yaxis',
                                         options=[{'label':i,'value':i} for i in columns],
                                         value= columns[0])],style={'width':'30%','display':'inline-block'}),
                    ]),
                ),], className="col-md-12 text-center"),
        
                     
         dbc.Row([
              dbc.Col(html.Div(
                  html.Div([
                      html.Div([dcc.Graph(id="chart")]),
                  ]),
              ), className="col-md-12 text-center")
            ], className="col-md-12 text-center",style={'backgroundColor': 'black','margin-right':'-15px',
                                                        'margin-left': '-15px'}),
         
                
     #footer
    dbc.Row([
        dbc.Col(html.Div(
            html.P([' DADI Soukaina Â© 2020. '], style={'color':'white', 'align':'center'})
                ), className="col-md-12 text-center"),
            ], style={'backgroundColor': 'black'}
        )
    ])


@app.callback(Output('chart', 'figure'),
              [ Input('xaxis','value'),
               Input('yaxis','value'),
               Input('plot_name','value')]
              )

def update_graph(xaxis_nm, yaxis_nm,plot_name):


    if plot_name  == 'Scatter-plot':
      return {'data':[go.Scatter(x=df[xaxis_nm],
                                y=df[yaxis_nm],
                                mode='markers')
                     ],

              'layaout':go.Layout(title='SCatter plot',
                                 xaxis={'title':xaxis_nm},
                                 yaxis={'title':yaxis_nm},
                                showlegend=True,
                               paper_bgcolor='black',
                               plot_bgcolor='black'
                                 
                                 ) 
               
             }
    elif plot_name == 'Bar-chart' :
      return {'data':[go.Bar(x=df[xaxis_nm],
                                y=df[yaxis_nm],
                                )
                     ],

              'layaout':go.Layout(title='bar plot',
                                 xaxis={'title':xaxis_nm},
                                 yaxis={'title':yaxis_nm},
                                showlegend=True,
                                paper_bgcolor='rgba(0,0,0,0)',
                                plot_bgcolor='rgba(0,0,0,0)')
               
             }
    elif plot_name == 'Line-chart':
       return {'data':[go.Scatter(x=df[xaxis_nm],
                                y=df[yaxis_nm],
                                mode='lines')
                     ],

              'layaout':go.Layout(title='SCatter plot',
                                 xaxis={'title':xaxis_nm},
                                 yaxis={'title':yaxis_nm},
                                 showlegend=True,
                                paper_bgcolor='rgba(0,0,0,0)',
                                plot_bgcolor='rgba(0,0,0,0)')
             }



if __name__ == '__main__':
    app.run_server(debug=False)
```
